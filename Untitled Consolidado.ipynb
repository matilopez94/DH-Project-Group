```python
#realizamos una lista que sera la indexacion de las nuevas columnas

index = ["1a","2a","3a","4a","5a","6a","7a","8a","9a","10a","11a","12a","13a","14a","15a"]


#creo las columnas a traves de un bucle y les asigno valor = 0
for i in range(0, len(index)):
    df_palermo[index[i]]=0 

#creo columna por valor "parrilla" y le asigno valor = 0
df_palermo['parrilla'] = 0

#creo columna por valor "precio" y le asigno valor = 0
df_palermo['precio'] = 0


#creo otra columna por valor "precio" para los casos del campo "title" y le asigno valor = 0

df_palermo['precio_title'] = 0

df_palermo.head(1)
```

```python
#creando indices

#Index.get_loc(key, method=None, tolerance=None) => get_loc extrae el nro referente al indice del item al que se lo aplica
#https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.Index.get_loc.html
index2 = np.zeros(15, dtype=int)

index_title = df_palermo.columns.get_loc("title")
index_description = df_palermo.columns.get_loc("description")

#creando extrayendo valor del indice de ambientes y los guardo en una lista
for i in range(0, len(index)):
    index2[i]= df_palermo.columns.get_loc(index[i])
    #print(index2[i])

#extrayendo valor del indice parrilla
index_parrilla = df_palermo.columns.get_loc("parrilla")

#extrayendo valor del indice precio
index_precio = df_palermo.columns.get_loc("precio")

index_precio_title = df_palermo.columns.get_loc("precio_title")


print(index_description, index2)
#23 para la descripcion, #columna rescatada
```

```python
#definir patron de dato

patron_1a = r'(1 AMB|1AMB|MONOAMB|ambiente unico)'
patron_2a = r'(2 AMB|2AMB})'
patron_3a = r'(3 AMB|3AMB)'
patron_4a = r'(4 AMB|4AMB)'
patron_5a = r'(5 AMB|5AMB)'
patron_6a = r'(6 AMB|6AMB)'
patron_7a = r'(7 AMB|7AMB)'
patron_8a = r'(8 AMB|8AMB)'
patron_9a = r'(9 AMB|9AMB)'
patron_10a = r'(10 AMB|10AMB)'
patron_11a = r'(11 AMB|11AMB)'
patron_12a = r'(12 AMB|12AMB)'
patron_13a = r'(13 AMB|13AMB)'
patron_14a = r'(14 AMB|14AMB)'
patron_15a = r'(15 AMB|15AMB)'
patron_parrilla = r'(parrilla)'
patron_precio = r'(USD|U\$D|\$)'

#creo lista de patrones
patrones = [ patron_1a, patron_2a, patron_3a, patron_4a, patron_5a, patron_6a, patron_7a, patron_8a, patron_9a, patron_10a, patron_11a, patron_12a, patron_13a, patron_14a, patron_15a]
```

```python
#aplicando el patron


#property DataFrame.iat(row, column)
#https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.iat.html

for row in range(0,len(df_palermo)):
    for i in range(0,len(patrones)):
        matcheo = re.search(patrones[i], df_palermo.iat[row, index_description], re.IGNORECASE)
        if matcheo != None:
            df_palermo.iat[row, index2[i]] = 1
    
# aplicando el patron para valor parrilla

for row in range(0, len(df_palermo)):
    matcheo = re.search(patron_parrilla, df_palermo.iat[row, index_description], re.IGNORECASE)
    if matcheo != None:
        df_palermo.iat[row, index_parrilla] = 1
        
for row in range(0, len(df_palermo)):
    matcheo = re.search(patron_precio, df_palermo.iat[row, index_description], re.IGNORECASE)
    if matcheo != None:
        df_palermo.iat[row, index_precio] = 1  
        
        
for row in range(0, len(df_palermo)):
    matcheo = re.search(patron_precio, df_palermo.iat[row, index_title], re.IGNORECASE)
    if matcheo != None:
        df_palermo.iat[row, index_precio_title] = 1 
```
